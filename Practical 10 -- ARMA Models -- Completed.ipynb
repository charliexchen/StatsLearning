Let's first generate some white noise and see what happens.

```r
n <- 100
time <- 1:n
set.seed(2018)
W <- rnorm(n, mean=0, sd = 0.2)
plot(time, W, type='l', ylim=c(-1,1))
```

```r
rho <- acf(W)
rho$acf
```

This is the empirical ACF (autocorrelatition). We expect there to be a peak at 1, and for none of them to be zero. The blue dashed bands are the 95% confidence intervals for uncorrelated white noise. The autocorrelation function $\rho(h)=\frac{\gamma(h)}{\gamma(0)}$. So we can compute the empirical autocovariance function $\hat\gamma(0)\hat\rho(h) = \hat\gamma(h)$.

```r
gamma<- rho$acf * var(W)
plot(gamma, type="b", xlab = 'Lag')
```

We can use this white noise to generate a AR(1) process.

```r
beta <- 0.9; sigma <- 0.2
set.seed(2018)
W <-rnorm(n, mean=0, sd=sigma)
Y <- rep(0, n)
Y[1] <- W[1] # initialise
for (t in 2:n) Y[t] <- beta * Y[t-1] + W[t]
plot(time, Y, type='l', col='blue', ylim=c(-1,1))
points(time, W, col='orange', type='l')
legend('topleft', c('Y', 'W'), col=c('blue','orange'), lty=1)
acf(Y)
```

This is not technically an AR(1) process, since the variance is not constant. We change the code slightly next to make it work properly. This doesn't really matter much, since the effect decays on a long chain. So for all intents and purpses this will work as AR(1).

The time series is correlated, so when the parameter is close to 1 the chain is highly regular. Naturally, if the parameter is close to 0 the series will be close to the white noise. If we set beta to be larger than 1, the chain explodes -- it is no longer causal. Negative $\beta$ gives oscillations due to the negative correlation.

```r
beta <- 0.9; sigma <- 0.2
set.seed(2018)
W <-rnorm(n, mean=0, sd=sigma)
Y <- rep(0, n)
Y[1] <- W[1] / sqrt(1-beta^2) # initialise
for (t in 2:n) Y[t] <- beta * Y[t-1] + W[t]
plot(time, Y, type='l', col='blue', ylim=c(-1,1))
points(time, W, col='orange', type='l')
legend('topleft', c('Y', 'W'), col=c('blue','orange'), lty=1)
acf(Y)
```

We now look at a MA(1) model.

```r
beta <- 0.9; sigma <- 0.2
set.seed(2018)
W <-rnorm(n, mean=0, sd=sigma)
Y <- rep(0, n)
Y[1] <- W[1] / sqrt(1-beta^2) # initialise
for (t in 2:n) Y[t] <- beta * W[t-1] + W[t]
plot(time, Y, type='l', col='blue', ylim=c(-1,1))
points(time, W, col='orange', type='l')
legend('topleft', c('Y', 'W'), col=c('blue','orange'), lty=1)
acf(Y)
```

There is a large correlation at lag 1, as expected for a MA(1) model.

```r
acf(arima.sim(list(order=c(0,0,1), ma=0.9), n=100, sd=0.2))
acf(arima.sim(list(order=c(0,0,1), ma=0.2), n=100, sd=0.2))
```

Now we try to simulate an ARMA(2,1) model. The parameters will be $\beta_1 = 0.9, \beta_2 = -0.3, \theta = 0.5$.

So $Y_n = 0.9 Y_{n-1} -0.3 Y_{n-2} +\epsilon_n + 0.5 \epsilon_{n -1}$, and $\epsilon \sim N(0, 0.2^2)$

```r
set.seed(2018)
Y <- arima.sim(list(order=c(2,0,1), ar=c(0.9, -0.3), ma=0.5), n=100, sd=0.2)
plot(Y, type='l')
acf(Y)
```

If an ARMA process is Causal, it can be expressed in the form: $Y_n = \sum_{i=0}^\infty \phi_i\epsilon_{n-i} $ There is a function to do this in R. We proved in lectures that this decays exponentially, so the cumulative plot converges.

```r
psi <- ARMAtoMA(ar=c(0.9,-0.3), ma=0.5, 100)
psi
plot(cumsum(abs(psi)))
```

There is built in method to convert ARMA to AR. However, we can still construct it ourselves in terms of ARMA to AR

```r
ARMAtoAR = function(ar, ma, lag.max){ARMAtoMA(ar = -ma, ma = -ar, lag.max)}
pi <- ARMAtoAR(ar=c(0.9,-0.3), ma=0.5, 100)
pi
```

We can now have a look at some real data.

```r
filePath <- "http://www.statslab.cam.ac.uk/~tw389/teaching/SLP18/data/"
fileName <- "wait"
dat <- read.table(paste0(filePath, fileName), header=T)
wait <- dat$wait
```

There's no immediate reason that this is wouldn't be nonstationary, but it is clear that this is not some MA(1) series -- the autocorrelation is significant even with lag 4.

```r
wait.ts <- ts(data=wait, start=c(1,1), frequency=24)
plot(wait.ts, xlab="Days")
acf(wait.ts)
```

Let's try to fit an AR(1) model to this data.

```r
wait.ar <-arima(wait.ts, order=c(1,0,0))
wait.ar
```

We now have some time series with non-zero mean. The intercept term is the MLE of the mean, so we have fitted: $Y_n-20.0841 =0.4724(Y_{n-1}-20.0841) + \epsilon_{n}$ 

Let's compare with an intercept only model -- this has much lower log likelihood and higher AIC.

Also, note that the Z-score (really a T-score, since the s.e. is also estimated) for $\phi$ is $0.4724/0.0676 = 6.98$, so it is significant and we reject the null.

```r
wait.ar <-arima(wait.ts, order=c(0,0,0))
wait.ar
```

Suppose we try to regress $Y_2,..., Y_n$ against $Y_1,..., Y_{n-1}$. This minimises:

$\sum_{i=0}^{n-1}(Y_{i+1} - \phi Y_i - \phi_0)^2$

However, our AR model is miminimising:

$l(\phi_i, \sigma^2, \mu| Y_1,..., Y_n ) = constant + \frac{\log(1-\phi^2)}{2} + \frac{1}{2\sigma^2}\left(\sum_{i=2}^n (Y_t-\phi_1Y_{t-1} - (1-\phi)\mu)^2+(1-\phi^2)(Y_1-c)^2\right)$

This is different, though close so we expect the regression coefficents to be different. Indeed, trying to fit it ourselves gives us:

```r
tmp<-data.frame(x=wait[-1], lag1=wait[-length(wait)])
lm(x~lag1, data=tmp)
```

We check the residuals to verify that the model is adequate. The residuals are approximately white noise, so we concluded we the solution is good.

```r
plot(wait.ar$residuals,type='l')
acf(wait.ar$residuals)
```

Now try a MA(4) model -- we use 4 because the ACF necessitates it:

```r
wait.ma <-arima(wait.ts, order=c(0,0,4))
wait.ma
```

```r
plot(wait.ma$residuals,type='l')
acf(wait.ma$residuals)
```

Again, the model appears to work well. It is often hard to to choose the correct model in the case of arma models. So in this case, we simply use the AIC to decide. The AIC chooses the MA model, since it has a smaller value.


Now, we take a look at series decomposition with some sales data:

```r
fileName <- "sales"
sales <- read.table(paste0(filePath, fileName), header=T)
head(sales)
```

```r
Sales <- as.numeric(unlist(sales[,-1]))
Sales
Year <- rep(colnames(sales)[-1], each = 12)
Month <- rep(sales[,1], 5)
time <- 1:length(Sales)
```

```r
plot(time,Sales)
```

There is a periodic trend in the data! So we cannot use arma directly. Instead, we try to for a linear model for the trend in order to extract the upward trend. Also, since this is a seasonal trend, we calculate the seasonal component by averaging the months across the years.

```r
trend <- lm(Sales ~ time)
d.Sales <- Sales - trend$fitted.values
seasonal <- rep(0, length(Sales))
for (k in 1:12){
# select the month in the time series
index <- which(Month==sales[k,1])
# the seasonal component is the average for that month
seasonal[index] <- mean(d.Sales[index])
}
residuals <- d.Sales - seasonal
old_par <- par(mfrow=c(3,1))
plot(time,trend$fitted.values,type='l',ylab="trend")
plot(time,seasonal,type='l',ylab="seasonal component")
plot(time,residuals,type='l',ylab="residuals")
par(old_par)
```

This gives us some residuals. Unfortunately, this still does not appear to be stationary, since the average shifts changes towards the end. Let's inspect the residuals

```r
plot(Sales)
points(seasonal+trend$fitted.values,type='l',col=4,lwd=2)
```

We can now see that this is not actually a good fit for more recent data.
