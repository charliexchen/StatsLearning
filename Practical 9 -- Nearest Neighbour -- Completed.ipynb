Try a kNN classifier:

```r
px1 <- seq(from = range(x[,1])[1], to = range(x[,1])[2], by = 0.1)
px2 <- seq(from = range(x[,2])[1], to = range(x[,2])[2], by = 0.1)

grid <- expand.grid(px1, px2)
plot(grid, pch='.')
points(grid, pch='.')
```

```r
pred <- knn(x, grid, y, k=5, prob=TRUE)
str(pred)
attr(pred, 'prob')
p <- ifelse(pred=='1', attr(pred, 'prob'), 1- attr(pred, 'prob'))
```

In order to visualise the data, we're going to classify a grid of dots and see where it gets us.

The mess of number above indicate the probaility that a point is going to be classed as one class. We can plot the decision boundaries where probability is 1/2.

```r
pmatrix <- matrix(p, length(px1), length(px2))
contour(px1, px2, pmatrix, levels=0.5, labels='', xlab='', ylab='', asp=1)
points(grid, pch='.', col=ifelse(p<0.5, 'orange', 'blue'))
points(x, pch=20, col=ifelse(y==0, 'orange', 'blue'))
```

By altering $k$, we get different boundaries. Smaller $k$ has lower bias but high variange and vice versa. Also, the larger values smooth the boundaries.

Let's try to find the optimal value using CV, which recommends $k=5$ or $k=13$.

```r
ks <- seq(1,99, by=2) # why by 2?
cv_err <- rep(0, length(ks))
for (k in ks){
j <- (k+1)/2
cv_err[j] <- sum(knn.cv(x,y,k) != y) / 200
}
plot(ks, cv_err, type='l')
```

```r
set.seed(2018)
m <- 100
pred <- matrix(0, dim(grid)[1], 9)
old_par <- par()
par(mfrow=c(3,3), mar=rep(0.5,4)) # mar sets margin
for (iter in 1:9){
ind <- sample(200, m)
pred[,iter] <- as.numeric(knn(x[ind,], grid, y[ind], k=1)) - 1
p <- matrix(pred[,iter], length(px1), length(px2))
# why is there no need to compute probability? Because it is a 1nn classifier
contour(px1, px2, p, levels=0.5, labels='', xlab='', ylab='', asp=1)
points(grid, pch='.', col=ifelse(p<0.5, 'orange', 'blue'))
points(x[ind,], pch=20, col=ifelse(y[ind]==0, 'orange', 'blue'))
}
par(old_par)
```

We can try to sample the data with replacement too visalise the amount of variation in the classifier. By combining the 9 datasets, we get a bagged sampler.

```r
p <- matrix(rowMeans(pred), length(px1), length(px2))
contour(px1, px2, p, levels=0.5, labels='', xlab='', ylab='', asp=1)
points(grid, pch='.', col=ifelse(p<0.5, 'orange', 'blue'))
points(x, pch=20, col=ifelse(y==0, 'orange', 'blue'))
```

This is still not great -- since we have sampled 100 out of 200 data points 9 times, the probability of a particular point being chosen is $1-(199/200)\times 100 \approx 0.394$. With $9$ bootstrap attempts, the probability of a point appearing at least 5 out 9 times is quite large (around 25%, in fact). This means outliers can still persist after bagging.

We can improve this by either increasing the number of times we bag or reducing the number of samples we draw (downsampling). Let's try again, this time sampling 40 instead of 100 -- this reduces the impact of immediate neightbours. As you can see, the boundaries are much nicer!

```r
set.seed(2018)
m <- 40
pred <- matrix(0, dim(grid)[1], 9)
old_par <- par()
par(mfrow=c(3,3), mar=rep(0.5,4)) # mar sets margin
for (iter in 1:9){
ind <- sample(200, m)
pred[,iter] <- as.numeric(knn(x[ind,], grid, y[ind], k=1)) - 1
p <- matrix(pred[,iter], length(px1), length(px2))
# why is there no need to compute probability? Because it is a 1nn classifier
contour(px1, px2, p, levels=0.5, labels='', xlab='', ylab='', asp=1)
points(grid, pch='.', col=ifelse(p<0.5, 'orange', 'blue'))
points(x[ind,], pch=20, col=ifelse(y[ind]==0, 'orange', 'blue'))
}
par(old_par)
p <- matrix(rowMeans(pred), length(px1), length(px2))
contour(px1, px2, p, levels=0.5, labels='', xlab='', ylab='', asp=1)
points(grid, pch='.', col=ifelse(p<0.5, 'orange', 'blue'))
points(x, pch=20, col=ifelse(y==0, 'orange', 'blue'))
```

Now for MNIST. You know the drill! First, import the data, call a kNN classifier with $k=9$ and have a look at the confusion matrix.

It's worked pretty well! The error rate is only 8%.

```r
filePath <- "http://www.statslab.cam.ac.uk/~tw389/teaching/SLP18/data/"
fileName <- "mnist.csv"
mnist <- read.csv(paste0(filePath, fileName), header = TRUE)
mnist$digit <- as.factor(mnist$digit)
visualise = function(vec, ...){ # function for graphically displaying a digit
image(matrix(as.numeric(vec),nrow=28)[,28:1], col=gray((255:0)/255), ...)
}
train <- mnist[1:4000, ]
test <- mnist[4001:6000, ]
pred = knn(train[,-1], test[,-1], train[,1], k=9)
actual <- test[,1]
sum(actual!=pred)/2000
table(actual, pred)
```

Little problem here -- Since we're using euclidean distance of images, we're not making use of information. We can try rotating the images to generate more data in order to explot this fact. Let's implement a function to do this and the generate some new data.

```r
rotate <- function(original, degree){
# rotate 28x28 greyscale images by the given degrees clockwise
rotated <- original
img <- as.matrix(original[,-1])
ind <- matrix(FALSE,40,40); ind[7:34,7:34] <- TRUE; ind <- as.vector(ind)
tmp <- matrix(0, dim(img)[1], 40*40); tmp[, ind] <- img
i <- rep(7:34, 28); j <- rep(7:34, each=28)
a <- degree/180*pi
i_rotated <- (i-20.5)*cos(a) + (j-20.5)*sin(a) + 20.5
j_rotated <- -(i-20.5)*sin(a) + (j-20.5)*cos(a) + 20.5
i0 <- floor(i_rotated); i1 <- ceiling(i_rotated); lambda_i <- i_rotated-i0
j0 <- floor(j_rotated); j1 <- ceiling(j_rotated); lambda_j <- j_rotated-j0
rotated[,-1] <- tmp[, i0+(j0-1)*40] * (1-lambda_i) * (1-lambda_j) +
tmp[, i0+(j1-1)*40] * (1-lambda_i) * lambda_j +
tmp[, i1+(j0-1)*40] * lambda_i * (1-lambda_j) +
tmp[, i1+(j1-1)*40] * lambda_i * lambda_j
return(rotated)
}
```

```r
train10c <- rotate(train, 10)
train20c <- rotate(train, 20)
train10a <- rotate(train, -10)
train20a <- rotate(train, -20)
par(mfrow=c(1,5), mar=c(0.5,0.5,4,0.5)) # mar sets margin
visualise(train20a[1,-1], main = 'deg = -20', asp=1)
visualise(train10a[1,-1], main = 'deg = -10', asp=1)
visualise(train[1,-1], main = 'deg = 0', asp=1)
visualise(train10c[1,-1], main = 'deg = 10', asp=1)
visualise(train20c[1,-1], main = 'deg = 20', asp=1)
```

Now we have generated some new data, let's have a go at the kNN again. The error is now lower at 5.6%!

```r
train_aug = rbind(train, train10c, train20c, train10a, train20a)
pred = knn(train_aug[,-1], test[,-1], train_aug[,1], k=9)
actual <- test[,1]
sum(actual!=pred)/2000
table(actual, pred)
```

This is equivalent to the previous classifier if we used a different distance. Let $R$ be a rotation by 10 degrees. We are effectively classifying using $d(x,x')=\min_{\{k=-2,-1,0,1,2\}} ||R^kx - x'||^2_2$.

Finally, some misclassified images -- these seem like reasonable misclassifications, since the images appear to be rather messed up.
