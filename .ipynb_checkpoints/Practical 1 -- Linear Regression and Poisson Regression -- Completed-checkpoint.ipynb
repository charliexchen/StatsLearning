```r
filePath <- "http://www.statslab.cam.ac.uk/~tw389/teaching/SLP18/data/"
fileName <- "housing.csv"
housing <- read.csv(paste0(filePath,fileName))
head(housing)
```

```r
cor(housing)
```

```r
pairs(housing)
```

```r
housing.lm1 <- lm(medv ~ crim + nox + rm + age + ptratio, data = housing)
summary(housing.lm1)
```

We can compute these values ourselves.

For the least squares, this is simply: $(X^TX)^{-1} X^TY$

Note that for the variance, we have: $\frac{1}{n-p}||Y-X\hat \beta||^2$

This is the unbiased estimator instead of the MLE.

variances of the estimators are $(X^TX)^{-1}\hat \sigma^2$




```r
X <- model.matrix (~crim + nox +rm +age +ptratio , data = housing)
Y <- housing$medv
betahat <- solve(t(X)%*%X, t(X)%*%Y)
Yhat <- X %*% betahat
RSS <- norm(Y-Yhat, type = "F")^2
sigmahat <- sqrt(RSS/(length(Y) - length(betahat)))
head(sigmahat)
sqrt(diag(solve (t(X)%*%X)))*sigmahat
```

The F-statistic is a test for the intercept only null model. Small p value indcates that we accept the alternative hypothesis of your current model.

```r
old_par <- par(mfrow=c(2,2))
plot(housing.lm1, which = c(1,2,3,4))
par(old_par)
```

1 -- fitted vs residuals. We should see little to no trend here, since we expect the noise between individual samples to be uncorrelated. This plot may also let us see if nonlinear trends are present. That might motivate a corresponding non-linear model.

2 -- we expect the resituals to behave with a normal distribution. We can verify this by normalising the variances and comparing to a standard normal distribution on a Q-Q plot. ($\hat \epsilon \sim N(0, \sigma^2 (I_n - P))$ where $P = X(X^TX)^{-1}X^T$. So we're plotting the quartiles of $\frac{\hat\epsilon}{\sqrt{1-P_{ii}}}$

3 -- Due to the homoscadesticity condition we expect that variance of the residuals does not depend on $Y$. So the plot between the sqrt of absolute values of residuals should not display any strong trends when plotted with $Y$. 

4 -- The Cook's distance is the impact of a particular datapoint on the overall plot (approximately the change in the log likelihood if a point is removed) . A large value may indicate outliers.

```r
old_par <- par (mfrow = c(3,3))
plot (housing.lm1, which=2)
for (iter in 1:8) {
    eps = rnorm (dim(housing)[1]) * summary (housing.lm1)$sigma
    
    plot (lm(fitted.values(housing.lm1) + eps ~ crim + nox + rm + age + ptratio, data = housing), which=2) 
    
}
```

The above plot generates synthetic data and the does a Q-Q plot the real data. There's no clear distinction between the real and fake data, so there is no clear abnormality. Now we shall look at some insurance data.

```r
fileName <- "insurance"
insurance <- read.table (paste0 (filePath, fileName), header = TRUE)
head(insurance)
with(insurance, plot(age, nclaims))
```

```r
ins.glm1 <- glm(nclaims ~ age, family = poisson, data = insurance)
ins.glm2 <- glm(nclaims ~ 1, family = poisson, data =insurance)
summary(ins.glm1)

anova(ins.glm2, ins.glm1, test = "Chisq")
```


Note that the residual deviance is not massive compaired to a $\chi^2_{33}$ distribution. Hence this is reasonable.

Also, the inclusion of $\beta$ decreases the deviance by $6.76$, which is significant given $\chi^2_1$.

The optimsation method used is Newton-Ralphson -- also known as Fisher scoring in this context.

During regression, we want to find the root to the score function (we know this is unique if it exists, since we have used the canonical link function, which ensures strict convexity).

So we have $\nabla l(b) = \sum_{i=1}^n x_i (Y_i - e^{x_i^T\hat \beta}) = 0$

We start with $\beta^{(0)} = 0$. 

For $t=1,2...$, we assign $\beta^{(t+1)} \leftarrow  \beta^{(t)} - (\nabla^2 l(\beta^{(t)}))^{-1} \nabla l(\beta^{(t)})$

```r
X <- model.matrix(~age, data=insurance)
Y <- insurance$nclaims
betahat <- rep(0,2)
for (iter in 1:5) {
    #We have the closed for the both the hessian and the derivative of the score function, so we just use that
    score <- t(X)%*%(Y - exp(X%*%betahat))
    infomat = -t(X)%*%diag(as.vector(exp(X%*%betahat)))%*%X
    betahat <- betahat - solve(infomat, score)
}
betahat
```
