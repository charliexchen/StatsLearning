Let's import the data first.

```r
filePath <- "http://www.statslab.cam.ac.uk/~tw389/teaching/SLP18/data/"
fileName <- "orthodont"
orth <- read.table(paste0(filePath, fileName), header = T)
head(orth)
# plot data, 'alpha' sets transparency level to display overlapping points
plot(orth$age, orth$distance, pch=20, col=rgb(0,0,0,alpha=0.4))
# plot individual growth curves
for (i in 1:30) {
ind <- orth$pid==paste0("P",i)
points(orth$age[ind], orth$distance[ind],type="l", col=i)
}
```

```r
orth.lm1 <- lm(distance ~ I(age-8), data = orth)
summary(orth.lm1)
plot(orth.lm1, which = c(1,2,3,4))
```

There's no clear trends in the residuals, bad fit in the QQ plot, change in variance in scale location or outliers in the cook's distance. In other words... These diagnostic plots don't necessarily mean a model is good!

Let's try a different model, where $Y_{ij} = \beta_0+\beta X_i +b_i+\epsilon_{ij}$, where $b_i\sim N(0, \tau^2)$ and $\epsilon_{ij} \sim N(0, \sigma^2)$.

This takes into account the fact that different people might have different intercepts.

```r
library(lme4)
orth.lme1 <- lmer(distance ~ I(age - 8) + (1 | pid), data=orth)
summary(orth.lme1)
```

This fits a REML model by default. The 1|PID means that the mixed effect occurs at the intercept level.

Let's try again using MLE instead of REML:

```r
orth.lme1b <- lmer(distance ~ I(age - 8) + (1 | pid), data=orth,
REML=FALSE)
summary(orth.lme1b)

```

We have obtained pretty much the same values for everything, but the variances parameters, in addition to the standard errors are a little smaller.


REML projects the data into the orthogonal complement of the column space of $X$ before estimating, so it is unbiased unlike MLE. This trade off is the reason for the difference.

This models is nested with the linear model, so we can try a hypothesis test. The value is very very large, and whilst it is inappropriate to compare using $\chi^2_1$, such a large value indicate that the mixed effects are important.

```r
lrstat <- 2*(logLik(orth.lme1b)[1] - logLik(orth.lm1)[1])
lrstat
```

Why is the $\chi_1^2$ invalid? Let's generate some random test statistics to check:

```r
set.seed(2018) # so that we all get the same 'random' numbers
B <- 100; lrstatMC <- rep(0,B)
attach(orth)
for (b in 1:B){
distanceMC <- 24.3 + (age - 8) * 0.497 + rnorm(30*4, sd=0.522)
boot.lm <- lm(distanceMC ~ I(age - 8))
boot.lme <- lmer(distanceMC ~ I(age - 8) + (1 | pid), REML=FALSE)
lrstatMC[b] <- 2*(logLik(boot.lme)[1] - logLik(boot.lm)[1])
}
lrstatMC
sum(lrstatMC<1e-10)
```

We have 62 zeroes! This means it can't be a chi square distributed. This means, in order to test the hypothesis, we have to use parametric bootstrap.

Since our statistic is so large, the bootstrap p-value is essentially 0. So indeed, we reject the null and take on the mixed model.


```r
boot.pval <- sum(lrstatMC>lrstat)/B
boot.pval
```

Let's fit a model with both random intercept and slope. This means the model is of the form:

$Y_{ij} = (\beta + b_{i1})X_i+b_{i0}+\epsilon_{ij}$

Where $\begin{bmatrix}b_{i0}\\b_{i1}\end{bmatrix} \sim N\left(\begin{bmatrix}0\\0\end{bmatrix} , \begin{bmatrix} \sigma^2_0, \sigma_1\sigma_0\rho\\\sigma_1\sigma_0\rho,\sigma^2_1 \end{bmatrix}\right)$ 

```r
orth.lme2 <- lmer(distance ~ I(age - 8) + (I(age - 8) | pid), data=orth)
summary(orth.lme2)
```

We conduct a random bootstrap -- the p-value is 0.88, so we don't reject the null hypothesis of the mixed intecept only model.

Now let's try doing this with some word frequency data with a general linear mixed model.

We don't have a nice way to do REML with generalised linear models, so that is not implemented (this instead uses MLE).

```r
library(languageR)
data(writtenVariationLijk)
head(writtenVariationLijk)
lijk.lme1 <- glmer(Count ~ Country + Register + (1|Word),
data = writtenVariationLijk, family = "poisson")
summary(lijk.lme1)
```

```r
The anova function compares two models using their likelihood ratio test
statistic. 
    
    Are the two tests valid? Both tests are for nested models. The first test is valid
since the additional parameter is in the fixed effect, and the null parameter of 0 is in the
interior of the parameter space. The second test is not valid since we are testing whether
separate random effects are needed for the two countries. The null parameter is hence on
the boundary of the parameter space and the chi-square distributional asymptotic is violated.
We should use a parametric bootstrap test in this case. However, since we observe
an extremely large likelihood ratio test statistic and parametric bootstrap test statistic is
typically smaller than the chi-square null distribution (because the bootstrap estimate of
the log-likelihood ratio is often precisely zero), we would expect the parametric bootstrap
test to produce a p-value very near 0 as well
```
