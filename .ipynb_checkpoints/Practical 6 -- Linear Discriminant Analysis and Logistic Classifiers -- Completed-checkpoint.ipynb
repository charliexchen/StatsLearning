```r
library (MASS)
data(iris)
head(iris)
pairs(iris[,1:4])
```

We try a Linear Discriminant classifier first. 

```r
iris.lda <- lda (Species~., data = iris)
iris.lda
```

```r
plot(iris.lda)
```

This plot is two dimensional, despite the fact there are three classes. This is because after sphering, we project the classes into an affine space spanned by the class centroids. 

We can do this manually, by

```r
A <- iris.lda$scaling
X <- as.matrix(iris[,1:4])
plot(X%*%A, pch=20, col=iris$Species)
pred <- predict(iris.lda, newdata=iris) # another way to obtain projected points
plot(pred$x, pch=20,col=iris$Species) # predict also centres the data
```

Let's have a look at the cross validation error -- it's not bad at all!

```r
sum(pred$class != iris$Species)/150 # training err
cvfit <- NULL
for (i in 1:150){
cvfit[i] <- predict(lda(Species~., data=iris[-i,]), newdata=iris[i,])$class
}
sum(cvfit != as.numeric(iris$Species))/150 # cross-validated err
```

Now let's have a look at a logistic classifier.

```r
library(nnet)
iris.logit <- multinom(Species~., data=iris, maxit=200)
coef <- t(coef(iris.logit))
coef
```

We classify to the class with the largest $\beta_i^Tx$, so we to clasify to, say Virginica, we need $\beta_\text{virg}^Tx>0$ and $\beta_\text{virg}^Tx>\beta_\text{versi}^Tx$. 

$\beta_\text{set}=0$ in this case. 

The prediction error is 1.3%, which is also pretty good.

```r
sum(predict(iris.logit, newdata=iris) != iris$Species)/150
```

We try to implement our own stochastic gradient descent here. The learning rate decays in order to allow for convergence. 

In fact, optimisation theory says that if learning rate $\alpha_i$ satisfy $\alpha_i\rightarrow 0$, $\sum_i \alpha_i = \infty$ and $\sum_i \alpha_i^2< \infty$, then it will converge.

```r
X <- model.matrix(Species~., data=iris)
y <- model.matrix(~Species, data=iris)[,-1] # indicator vectors for labels
labels <- levels(iris$Species)
n <- dim(X)[1]; p <- dim(X)[2] # num of obs and num of covariates
expit = function(v) exp(v)/(sum(exp(v))+1) # generalised expit for multinomial
logL <- function(beta, X, y){ # loglikelihood of the coefficients
sum(log(exp(rowSums(y*(X%*%beta)))/(rowSums(exp(X%*%beta))+1)))
}
beta <- matrix(0, p, 2) # initialise matrix of linear coefficients
nepoch = 200
set.seed(1122); shuffle = sample(150) # randomly shuffle indices
for (epoch in 1:nepoch){
# stochastic gradient descent updates
for (i in shuffle){
alpha <- 1/(1+epoch/10) # step size (learning rate)
xi = X[i, ,drop=T]; yi = y[i, ,drop=T]
prob <- expit(t(beta)%*%xi) # predicted probs of i-th obs with current betas
beta <- beta + alpha * xi %*% t(yi - prob) # SGD update
}
cat('epoch = ', epoch, 'loglik = ', logL(beta, X, y), '\n')
}
```

```r
beta
coef
logL(beta, X,y)
logL(coef,X,y)
```

We have obtained very different coeffiencts from Newton Ralphson and Stochastic Gradient descent. The stocastic descent method's log likelihood is worse, since it is a lower order method.

Now let's try our good friend MNIST

```r
filePath <- "http://www.statslab.cam.ac.uk/~tw389/teaching/SLP18/data/"
fileName <- "mnist.csv"
mnist <- read.csv(paste0(filePath, fileName), header = TRUE)
mnist[1:10,1:10]
mnist$digit <- as.factor(mnist$digit)
visualise = function(vec, ...){ # function for graphically displaying a digit
image(matrix(as.numeric(vec),nrow=28)[,28:1], col=gray((255:0)/255), ...)
}
```

```r
old_par <- par(mfrow=c(2,2))
for (i in 1:4) visualise(mnist[i,-1])
par(old_par)
```

```r
train <- mnist[1:4000,]
identical <- apply(train, 2, function(v){all(v==v[1])})
train <- train[,!identical]
test <- mnist[4001:6000,!identical]
```

```r
mnist.lda <- lda(digit~., data=train)
pred <- predict(mnist.lda, test)
sum(pred$class != test$digit)/2000
```

```r
We have error 16.9%, which is... not great.
```

```r
plot(pred$x, pch=20, col=as.numeric(test$digit)+1,
xlim=c(-10,10), ylim=c(-10,10))
legend('topright', lty=1, col=1:10, legend=0:9)
```

```r
train.err <- test.err <- rep(0, 20)
for (r in 1:20){
train.err[r] <- sum(predict(mnist.lda, train, dim=r)$class != train$digit)/4000
test.err[r] <- sum(predict(mnist.lda, test, dim=r)$class != test$digit)/2000
}
plot(train.err,type='l', col='orange',
ylim=range(c(train.err,test.err)), ylab='error')

points(test.err,type='l', col='blue')
legend('topright', c('train err', 'test err'), col=c('orange', 'blue'), lty=1)
```

We can try to fit a reduced rank classifier, which only uses the first couple of most important cannonical directions. The first two cannonical directions already show some good separations between classes, so we plot between the rest to see how it improves.

We see that all useful information are essentially captured in the first ten canonical directions. This makes since since we have 10 classes and the classifier is linear.

Now let's try a Logistic Classifier -- again we use Stochastic gradient descent:

```r
mnist.logit <- multinom(digit~., data=train, MaxNWts=100000)
sum(predict(mnist.logit, newdata=test) != test$digit)/2000
```

```r
X <- cbind(1,as.matrix(train[,-1])/255)
X.test <- cbind(1,as.matrix(test[,-1])/255)
y <- model.matrix(~digit-1, data=train)[,-1] # use digit 0 as baseline
nepoch <- 100
beta <- matrix(0, 655, 9)
train.err <- test.err <- rep(0,nepoch)
set.seed(1122); shuffle = sample(4000) # randomly shuffle indices
for (epoch in 1:nepoch){
for (i in shuffle){
alpha <- 1/(1+epoch/10)
xi <- X[i, ,drop=T]; yi <- y[i, ,drop=T]
prob <- expit(t(beta)%*%xi)
beta <- beta + alpha * xi%*%t(yi - prob)
}
train.err[epoch] <- sum((0:9)[max.col(cbind(0,X%*%beta))] != train[,1])/4000
8
test.err[epoch] <- sum((0:9)[max.col(cbind(0,X.test%*%beta))] != test[,1])/2000
}
test.err[nepoch]
```

```r
plot(train.err, type='l', col='orange', ylim=range(c(train.err,test.err)),
xlab='epoch', ylab='error')
points(test.err, type='l', col='blue')
legend('topright', c('train err', 'test err'), col=c('orange', 'blue'), lty=1)
```

Note that stochastic gradient descent gave a lower test error, despite lower lower log likelihood, so it has acted as a sort as regularisation.

```r
pred <- (0:9)[max.col(cbind(0,X.test%*%beta))]
err_ind <- (1:2000)[pred != test[,1]]
old_par <- par(mfrow = c(3,3))
for (i in 1:9){
visualise(mnist[4000+err_ind[i],-1],
main=paste0('true=', test[err_ind[i],1], ', pred=', pred[err_ind[i]]))
}
par(old_par)
```
